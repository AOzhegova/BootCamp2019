### Problem 1

```python
import numpy as np
from matplotlib import pyplot as plt
import math
```

```python
def var_means_func(n):
    sample = np.random.normal(size = (n, n))
    m_rows = np.mean(sample, axis = 1)
    return(np.var(m_rows))
```

```python
var_means_func(2)
```

```python
res = []
for i in np.arange(100, 1100, 100):
    res.append(var_means_func(i))
print(res)
```

```python
plt.plot(res)
plt.show()
```

### Problem 4

```python
x = np.linspace(.1, 2, 200)
# Create a subplot to cover the left half of the figure.
ax1 = plt.subplot(121)
ax1.plot(x, np.exp(x), 'k', lw=2)
ax1.plot(x, np.exp(2*x), 'b', lw=2)
plt.title("Exponential", fontsize=18)
#Create another subplot to cover the right half of the figure
ax2 = plt.subplot(122)
ax2.plot(x, np.log(x), 'k', lw=2)
ax2.plot(x, np.log(2*x), 'b', lw=2)
ax2.set_title("Logarithmic", fontsize=18)
plt.show()
```

```python
x = np.linspace(-5, 5, 100)
# 1. Use plt.subplot() to switch the current.
plt.subplot(121)
plt.plot(x, 2*x)
plt.subplot(122)
plt.plot(x, x**2)
```

```python
# 2. Use plt.subplot() to explicitly grab the two subplot axes.
ax1 = plt.subplot(121)
ax1.plot(x, 2*x)
ax2 = plt.subplot(122)
ax2.plot(x, x**2)
```

```python
# 3. Use plt.subplots() to get the figure and all subplots simultaneously.
fig, axes = plt.subplots(1, 2)
axes[0].plot(x, 2*x)
axes[1].plot(x, x**2)
```

```python
x = np.linspace(0, 2 * np.pi, 100)
ax1 = plt.subplot(221)
ax1.plot(x, np.sin(x), '-', color = 'green')
ax1.axis([0, 2 * np.pi, -2, 2])
ax1.set_title('sin(x)')
ax2 = plt.subplot(222)
ax2.plot(x, np.sin(2*x), '--', color = 'red')
ax2.axis([0, 2 * np.pi, -2, 2])
ax2.set_title('sin(2x)')
ax3 = plt.subplot(223)
ax3.plot(x, 2 * np.sin(x), '--', color = 'blue')
ax3.axis([0, 2 * np.pi, -2, 2])
ax3.set_title('2*sin(x)')
ax4 = plt.subplot(224)
ax4.plot(x, 2 * np.sin(2*x), ':', color = 'magenta')
ax4.axis([0, 2 * np.pi, -2, 2])
ax4.set_title('2*sin(2x)')
plt.axis([0, 2 * np.pi, -2, 2])
plt.suptitle('Graphs');
```
