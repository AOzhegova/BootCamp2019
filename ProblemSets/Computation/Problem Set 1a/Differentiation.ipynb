### Problem 2

```python
#Calculate forward first-order quotient 
def forward_first_order(f, x, h):
    res = []
    for xi in x:
        resi = (f(xi + h) - f(xi))/h
        res = np.append(res, resi)
    return(res)
```

```python
#Calculate forward second-order quotient 
def forward_second_order(f, x, h):
    res = []
    for xi in x:
        resi = (-3 * f(xi) + 4 * f(xi + h) - f(xi + 2*h))/2*h
        res = np.append(res, resi)
    return(res)
```

```python
#Calculate backward first-order quotient 
def backward_first_order(f, x, h):
    res = []
    for xi in x:
        resi = (f(xi) - f(xi - h))/h
        res = np.append(res, resi)
    return(res)
```

```python
#Calculate backward second-order quotient 
def backward_second_order(f, x, h):
    res = []
    for xi in x:
        resi = (3* f(xi) - 4 * f(xi - h) + f(xi - 2 * h))/2 * h
        res = np.append(res, resi)
    return(res)
```

```python
#Calculate centered second-order quotient 
def centered_second_order(f, x, h):
    res = []
    for xi in x:
        resi = (f(xi + h) - f(xi - h))/2 * h
        res = np.append(res, resi)
    return(res)
```

```python
#Calculate centered fourth-order quotient 
def centered_fourth_order(f, x, h):
    res = []
    for xi in x:
        resi = (f(xi - 2 * h) - 8 * f(xi - h) + 8 * f(xi + h) - f(xi + 2 * h))/12 * h
        res = np.append(res, resi)
    return(res)
```

```python
#This is our function
def funny_function(x):
    return(sin(x) + 1)**(sin(cos(x)))
```

```python
x_array = np.linspace(-np.pi, np.pi, 20)
```

```python
y_f1 = forward_first_order(funny_function, x_array, 0.01)
y_f2 = forward_second_order(funny_function, x_array, 0.01)
y_b1 = backward_first_order(funny_function, x_array, 0.01)
y_b2 = backward_second_order(funny_function, x_array, 0.01)
y_c2 = centered_second_order(funny_function, x_array, 0.01)
y_c4 = centered_fourth_order(funny_function, x_array, 0.01)
```

```python
ax1 = plt.subplot(331)
ax1.plot(x_array, y_f1)
ax1.set_title('Forward first-order')
ax1.set_xlabel('x')
ax2 = plt.subplot(332)
ax2.plot(x_array, y_f2)
ax2.set_title('Forward second-order')
ax2.set_xlabel('x')
ax3 = plt.subplot(333)
ax3.plot(x_array, y_b1)
ax3.set_title('Backward first-order')
ax3.set_xlabel('x')
ax4 = plt.subplot(334)
ax4.plot(x_array, y_b2)
ax4.set_title('Backward second-order')
ax4.set_xlabel('x')
ax5 = plt.subplot(335)
ax5.plot(x_array, y_c2)
ax5.set_title('Centered second-order')
ax5.set_xlabel('x')
ax6 = plt.subplot(336)
ax6.plot(x_array, y_c4)
ax6.set_title('Centered fourth-order')
ax6.set_xlabel('x')

plt.show()
```

```python
def jacobian_approx(f, x0, h):
    e = np.eye(len(x0))
    jacobian = np.empty([2, len(x0)])
    x0 = np.array(x0)
    for i in range(0, len(x0), 1):
        jacobian[:, i] = (f(x0 + h * e[:, i]) -
                          f(x0 - h * e[:, i]))/(2*h)
    return jacobian
```

```python
import autograd.numpy as anp  # Thinly-wrapped numpy
from autograd import grad 
from autograd import elementwise_grad as egrad
```

```python
def T(x, n):
    if n == 0:
        return anp.ones_like(x)
    elif n == 1:
        return x
    else:
        return 2 * x * T(x, n - 1) - T(x, n - 2)
```

```python
def funny_function(x):
    return(sin(x) + 1)**(sin(cos(x)))
```

```python
T_prime = egrad(T)
x = anp.linspace(-1, 1)
```

```python
plt.figure(figsize = (10, 10))
for n in range(5):
    plt.plot(x, T_prime(x, n), label = f'n = {n}')
plt.legend()
plt.show()
```
